```python
import turtle as tu
from random import randrange



# Initialisation du jeu
ts = tu.getscreen()
ts.clear()
ts.bgpic("champcourse2.gif")

ts.title("Bienvenue Ã  la course des tortues !")
ts.setup (width=1400, height=800, startx=0, starty=0)



# DÃ©clarez les 5 tortues et positionnez-les sur leurs hexagones respectifs
def createTurtle(x,y, color):
    turtleCreate = tu.Turtle()
    turtleCreate.color(color)
    turtleCreate.penup()
    turtleCreate.goto(x,y)
    turtleCreate.seth(0)
    turtleCreate.shape('turtle')
    turtleCreate.speed(5)
    return turtleCreate


startY = -320
startX = -650

Michaelangelo = createTurtle(startX, startY, 'orange')
startY += 160
Leonardo = createTurtle(startX, startY, 'deep sky blue')
startY += 160
Raphael = createTurtle(startX, startY, 'red')
startY += 160
Donatello = createTurtle(startX, startY, 'purple')
startY += 160
Splinter = createTurtle(startX, startY, 'dark slate gray')

# Demander de saisir dans la console les prÃ©dictions des joeurus 1 et 2 dans le format 1,2,3
def askPredicate():
    print ("Enter your predication : ")
    x = input()
    predicate = x.split(',')
    if (len(predicate) == 3):
        return predicate


print ("J1")
predicateJ1 = askPredicate()

print ("J2")
predicateJ2 = askPredicate()

FinishX = 680
players = [Michaelangelo, Leonardo, Raphael, Donatello, Splinter]

# A l'aide d'une boucle while, faire courir les tortues en tirant un nombre entre 0 et 5 qui reprÃ©sente le nombre de pixels du dÃ©placement vers la droite
isFinish = False
classement = []

#while(isFinish == False):
while(len(players) != 0):
    for p in players: 
        p.forward(randrange(5))
        if(p.xcor() >= FinishX):
            isFinish = True
            players.remove(p)
            classement.append(p)

print ("End Course")
i = 1

for turt in classement:
    print (i + turt.color())
    i += 1
    
print (classement)
tu.exitonclick()

turtle_arbitre = turtle.Turtle()
turtle_arbitre.goto(0,0)
turtle_arbitre.color("Black")
turtle_arbitre.write("Le joueur 1 à gagné", move=True, align="left", font=("Arial", 16, "normal"))

turtle.mainloop()
```
